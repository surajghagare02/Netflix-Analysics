```python
subscription = df.groupby(['Country', 'Subscription Type'])['User ID'].count().unstack()
subscription.plot(kind='bar', stacked=False, figsize=(10, 6))
plt.title('Subscription Plan Counts by Country')
plt.xlabel('Country')
plt.ylabel('Number of Subscriptions')
plt.xticks(rotation=45)
plt.legend(title='Subscription Type')
plt.tight_layout()
plt.show()
```

```python
netflix_data = pd.crosstab(df['Device'], df['Subscription Type'])
netflix_data.plot(kind='bar', stacked=True)
plt.title('Device Usage by Subscription Type')
plt.xlabel('Device')
plt.ylabel('Count')
plt.show()
```

```python
df_age = df.groupby('Age')['Monthly Revenue'].agg(['sum', 'count'])
df_age = df_age.reset_index()
df_age = df_age.rename(columns={'sum': 'Monthly Revenue', 'count': 'User Count'})
df_age.head()
```

```python
age = df_age['Age']
revenue_by_age = df_age['Monthly Revenue']
count_by_age = df_age['User Count']


# Premier sous-graphique
plt.plot(age, revenue_by_age)
plt.title('Revenu Mensuel par Age')
plt.xlabel('Age')
plt.ylabel('Revenu mensuel')

plt.show()
```
